```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('D:/aiml/Cars93.csv')
df
```

```python
fig=plt.figure(figsize=[3,3])
ax=fig.add_axes([0,0,1,1])
ax.hist(df['MPG.city'])
ax.set_title('histogram to check frequency of mpg')
ax.set_ylabel('miles per gallon',fontsize='12')
ax.set_xlabel("frequency",fontsize='12')
fig.show()
```

```python
fig=plt.figure(figsize=[3,3])
ax=fig.add_axes([0,0,1,1])
ax.set_title('to determine whether a car with high horsepower gives lower.mileage')
ax.set_ylabel('Horsepower',fontsize='12')
ax.set_xlabel("MPG",fontsize='12')
ax.scatter(df['Horsepower'],df['MPG.city'])
fig.show()
```

```python
fig=plt.figure(figsize=[3,3])
ax=fig.add_axes([0,0,1,1])
ax.set_title('price range of all different cars')
ax.set_ylabel('Horsepower',fontsize='12')
ax.set_xlabel("MPG",fontsize='12')
ax.boxplot(df['Price'])
fig.show()
```

```python
line_df=df.groupby(by=['Horsepower'])
line_df=line_df.mean()
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[3,3])
ax=fig.add_axes([0,0,1,1])
plt.plot(df['Horsepower'],df['EngineSize'],label='EngineSize')
ax.set_title('variation in EngineSize')
ax.set_ylabel('EngineSize')
ax.set_xlabel('Horsepower')
ax.legend()
fig.show()
```
