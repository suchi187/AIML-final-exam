```python
cat_features = ['batting_team','bowling_team','batsman','non_striker','bowler']
df = pd.get_dummies(df,columns=cat_features)
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler
# finding the mean and std deviation of numerical columns
scaler=StandardScaler().fit(df[['match_id','inning','over','ball','is_super_over','wide_runs','bye_runs','legbye_runs','noball_runs','penalty_runs','batsman_runs','extra_runs','total_runs']]) 
# scaling columns to a common range
df_normalized = scaler.transform(df[['match_id','inning','over','ball','is_super_over','wide_runs','bye_runs','legbye_runs','noball_runs','penalty_runs','batsman_runs','extra_runs','total_runs']])
#note that output is a 2-Dimensional array of normalized data
df_normalized
```
